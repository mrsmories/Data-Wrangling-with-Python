```python
df['Membership end date'] = pd.to_datetime(df['Membership end date'], errors='coerce')
df['Member type'] = df['Member type'].astype('category')
df.info()
```

```python
# Select all columns except 'Street' and 'Apt/Ste/Unit' and assign back to df
columns_to_keep = [col for col in df.columns if col not in ['Street', 'Apt/Ste/Unit']]
df = df[columns_to_keep]
df.head()
```

```python
#Rename 'New Address' to 'Address'
df = df.rename(columns={'New Address': 'Address'})
df.head()
```

```python
df.columns = df.columns.str.strip()
df.head()
```

```python
#clean the 'Member ID' column string data
df['Member ID'] = df['Member ID'].astype(str).str.strip().str.lower()
df['Member ID'] = df['Member ID'].str.replace(r'\s+', '', regex=True)
df
```

```python
# Generate random ages between 1 and 90
df['Age'] = np.random.randint(1, 91, size=len(df))
df.head()
```

```python
# Extract patterns from the City column
# This will handle various formats like:
# "08029 Barcelona" -> zip: "08029", city: "Barcelona"
# "D-79713 Bad Saeckingen" -> zip: "D-79713", city: "Bad Saeckingen"
# "6644 AZ Ewijk" -> zip: "6644 AZ", city: "Ewijk"

# Pattern explanation:
# ^([A-Z]?-?\d+\s*[A-Z]{0,2})\s+(.+)$
# - ^ start of string
# - ([A-Z]?-?\d+\s*[A-Z]{0,2}) captures zip codes (optional letter, optional dash, numbers, optional space, optional 2 letters)
# - \s+ one or more spaces
# - (.+) captures the rest as city name
# - $ end of string

extracted = df['City'].str.extract(r'^([A-Z]?-?\d+\s*[A-Z]{0,2})\s+(.+)$')

# Find rows where extracted a pattern AND Zip Code is currently empty/null
mask = extracted[0].notna() & (df['Zip Code'].isna() | (df['Zip Code'] == ''))

# Move the extracted zip codes to the Zip Code column
df.loc[mask, 'Zip Code'] = extracted.loc[mask, 0]

# Clean the City column
df.loc[mask, 'City'] = extracted.loc[mask, 1]

print("Rows that were updated:")
print(df.loc[mask, ['City', 'Zip Code']])
```
